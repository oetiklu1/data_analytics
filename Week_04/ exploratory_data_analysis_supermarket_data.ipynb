```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('supermarkets_data_enriched.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
import matplotlib.pyplot as plt

# Count the number of supermarkets per brand
brand_counts = df['brand'].value_counts()

# Create a barchart with the number of supermarkets per brand
plt.figure(figsize=(10, 6))
brand_counts.plot(kind='bar')
plt.title('Number of Supermarkets per Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Supermarkets')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

# Create a scatterplot matrix of the numeric variables
numeric_vars = ['lat', 'lon', 'pop', 'pop_dens', 'frg_pct', 'emp']
sns.pairplot(df[numeric_vars])
plt.show()
```

```python
# Filter the dataframe for specific brands
brands_to_plot = ['Denner', 'Volg', 'Landi']
df_filtered = df[df['brand'].isin(brands_to_plot)]

# Create a plot with the locations of supermarkets in different colors according to their brand
plt.figure(figsize=(10, 6))
for brand in brands_to_plot:
    subset = df_filtered[df_filtered['brand'] == brand]
    plt.scatter(subset['lon'], subset['lat'], label=brand, alpha=0.6)

plt.title('Locations of Supermarkets by Brand')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```
